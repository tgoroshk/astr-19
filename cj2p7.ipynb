Create a Jupyter notebook, 
Import matplotlib. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

Write cells that create an array x ranging from [0,1] in 100 steps and that defines a function that returns exp(x).  

```python
x = np.linspace(0,1,100)
```

```python
def exp_of_x(input_array):
    return np.exp(input_array)
```

In a new cell use the function to set y=exp(x), 
and then plot x vs. y.
Label the x-axis as “Time [milliseconds]” and the y-axis as “Awesomeness”. 
Save the figure as a PDF.

```python

y = exp_of_x(x)

plt.figure()
plt.plot(x,y)

plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")

plt.title("Code Journal 2 - prompt #7")
plt.show()

plt.savefig("a19_cj2p7.pdf")
```
