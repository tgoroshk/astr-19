```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng=np.random.default_rng()
lam = 10 # must be greater than 0
random_data = rng.poisson(lam=lam,size=1000) #size = single value if lam is a scalar

fig, ax = plt.subplots(figsize=(5, 5))

ax.hist(random_data,bins=100)

ax.set_xlabel("Number of Events (k)")
ax.set_ylabel("Frequency (number of entries per bin)")
ax.set_title("Histogram of Poisson Random Numbers (lambda=10)")

plt.savefig("a19_cj2p10.pdf")
plt.show()
```
